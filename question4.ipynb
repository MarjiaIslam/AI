```python
import numpy as np 
import matplotlib.pyplot as plt
import math
x = np.arange(0, 3.1416, 0.01)
y = np.arange(math.pi, 2*3.1416, 0.01)

a = np.sin(x)
b = np.cos(y)
plt.plot(x, a, "red")
plt.plot(y, b, "blue")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Sine & cosine graph")
plt.legend("sin")
plt.legend("cos")
plt.show()
```
